# Gestion des colonnes avec des jours négatifs pour les convertir en années

```python
def convert_to_year(x,is_negative=False):
    x=x/365
    if(is_negative):
        x=x*(-1)
    return x

#We assume that the absurd value given to DAYS_EMPLOYED (environ +365000) was to protect retired people from being misinterpretated as pure jobless people.

df["DAYS_EMPLOYED_FIX"]=df["DAYS_EMPLOYED"].apply(lambda x: x if x<0 else 0)
df["AGE"]=df["DAYS_BIRTH"].apply(lambda x: convert_to_year(x,is_negative=True))
df["AGE_WORK"]=df["DAYS_EMPLOYED_FIX"].apply(lambda x: convert_to_year(x,is_negative=True))

df=df.drop(columns=["DAYS_EMPLOYED_FIX","DAYS_BIRTH","DAYS_EMPLOYED"])
```

<H2> Gestion de la table Bureau</H2>

```python
data_2 = pd.read_csv("data/bureau.csv")
data_2 = data_2[["SK_ID_CURR", "CREDIT_ACTIVE", "DAYS_CREDIT", "DAYS_CREDIT_ENDDATE", "CNT_CREDIT_PROLONG", "AMT_CREDIT_SUM", "AMT_CREDIT_SUM_DEBT", "AMT_CREDIT_SUM_OVERDUE"]]
data_2 = data_2[data_2['CREDIT_ACTIVE'] == 'Active']

aggregations = {
    'CREDIT_ACTIVE': lambda x: x.value_counts(),
    'DAYS_CREDIT': 'min',
    'DAYS_CREDIT_ENDDATE': 'max',
    'CNT_CREDIT_PROLONG': 'count',
    'AMT_CREDIT_SUM': 'sum',
    'AMT_CREDIT_SUM_DEBT': 'sum',
    'AMT_CREDIT_SUM_OVERDUE':'sum'
}

# Appliquer les fonctions d'agrégation par groupe (ID)
resultat = data_2.groupby('SK_ID_CURR').agg(aggregations).reset_index()

resultat.columns = ['SK_ID_CURR','CREDIT_ACTIVE_counts','DAYS_CREDIT_min','DAYS_CREDIT_ENDDATE_MAX','CNT_CREDIT_PROLONG_count','AMT_CREDIT_SUM_active','AMT_CREDIT_SUM_DEBT_active', 'AMT_CREDIT_SUM_OVERDUE_sum']
```

```python
# Left join on df
df_new = pd.merge(df, resultat, on='SK_ID_CURR', how='left')
df_new = df_new.drop_duplicates(subset=['SK_ID_CURR'])
```
